# **Ahora que ya manejamos la introdución de datos**

---


Pasamos a realizar las operaciones. Para ello crearemos funciones pequeñas que sólo realizan lo que haríamos en una hoja, evaluar `x` en la funcion, su derivada, calcular la ordenada al origen de la recta tangente e y finalmente aplicamos Newton-Raphson para hallar la raíz de la recta tangente que nos dará un nuevo `x`.

$$x_{n+1} = \frac{-b}{f'(x)}$$


<br>

````python
def f(x):
    return a*x**3 + b*x**2 + c*x + d

def derivada(x):
    return 3*a*x**2 + 2*b*x + c

def ordenada(x):
    return f(x) - derivada(x) * x

def raiz(x_ini, max_i):
  import time
  x=x_ini
  for _ in range(max_i):
    if derivada(x) == 0:
      print("¡Derivada igual a cero! Reintentar con otro x inicial")
      return None
    x = -ordenada(x) / derivada(x)
    print(x)
    time.sleep(1)
  return x
````

<br>

````python
newton=raiz(n,i)
print("\n")
display(Math(fr"\text{{La raíz es aproximadamente: }} \textbf{{x}} \approx {newton} \text{{, en }} {i} \text{{ iteraciones.}}"))
````

# Mostrar una Tabla:

---

Con las iteraciones, se agrega otro método de la clase display: [clear_output](https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html#IPython.display.clear_output) se usa [time.sleep(secs)](https://docs.python.org/3/library/time.html#time.sleep) para que vayan mostrandose las iteracciones cada un segundo. Se usa LaTeX para el armado de la tabla mediante un `\begin{array}`.

<br>


````python
from IPython.display import display, Math, clear_output
import time

def raiz(x_ini, max_i):
    x = x_ini
    # Encabezado de la tabla (solo se muestra una vez)
    tabla = r"""
    \begin{array}{|c|c|c|c|c|c|}
    \hline
    \text{N° Iteración} & x & f(x) & \textbf{Pendiente en }(x, f(x) & \textbf{Ordenada de la tangente} & \textbf{Raíz de la tangente} \\
    \hline
    """
    display(Math(tabla + r"\hline \end{array}"))  # Tabla inicial vacía

    for _ in range(max_i):
        dx = derivada(x)
        if dx == 0:
            clear_output(wait=True)
            display(Math(r"\text{¡Error! Derivada cero.}"))
            return None

        x_nuevo = -ordenada(x) / dx
        tabla += fr"{_ + 1} & {x:} & {f(x):} & {dx:} & {ordenada(x):} & {x_nuevo:} \\ \hline "

        clear_output(wait=True)  # ¡Borra la salida anterior!
        display(Math(tabla + r"\end{array}"))  # Muestra la tabla actualizada
        time.sleep(1)  # Pausa de 1 segundo
        x = x_nuevo

    return x
````

<br>

````python
newton=raiz(n,i)
print("\n")
display(Math(fr"\text{{La raíz es aproximadamente: }} \textbf{{x}} \approx {newton} \text{{, en }} {i} \text{{ iteraciones.}}"))
````

````python
from dash import Dash, html, dcc, dash_table, Input, Output, State
import dash
import plotly.graph_objs as go
import numpy as np

app = Dash(__name__)

a = b = c = d = n = i = 0
funcion = ""

app.layout = html.Div([
    html.H3("Método de Newton-Raphson"),
    html.Div([
        html.Div("f(x) = (a):", style={'display': 'inline-block', 'width': '80px'}),
        dcc.Input(id='input-a', type='number', value=1, style={'width': '60px', 'display': 'inline-block', 'marginRight': '5px'}),
        html.Span("x³ +", style={'display': 'inline-block', 'width': '40px'}),

        dcc.Input(id='input-b', type='number', value=0, style={'width': '60px', 'display': 'inline-block', 'marginRight': '5px'}),
        html.Span("x² +", style={'display': 'inline-block', 'width': '40px'}),

        dcc.Input(id='input-c', type='number', value=0, style={'width': '60px', 'display': 'inline-block', 'marginRight': '5px'}),
        html.Span("x +", style={'display': 'inline-block', 'width': '40px'}),

        dcc.Input(id='input-d', type='number', value=0, style={'width': '60px', 'display': 'inline-block'})
        ], style={'marginBottom': '15px'}),

       html.Div([
            html.Label("Valor inicial (x₀):", style={'display': 'inline-block', 'width': '150px'}),
            dcc.Input(id='input-n', type='number', value=2, style={'width': '80px', 'display': 'inline-block'})
        ], style={'marginBottom': '10px'}),

        html.Div([
            html.Label("Número de iteraciones:", style={'display': 'inline-block', 'width': '150px'}),
            dcc.Input(id='input-i', type='number', value=9, style={'width': '80px', 'display': 'inline-block'})
        ], style={'marginBottom': '15px'}),

        html.Button("Iniciar", id="start-btn", n_clicks=0),

        html.Div(id='funcion-math'),

    dash_table.DataTable(
        id="tabla-newton",
        columns=[
            {"name": "Iteración", "id": "iterador"},
            {"name": "x", "id": "x"},
            {"name": "f(x)", "id": "fx"},
            {"name": "Pendiente en (x, f(x))", "id": "derivada"},
            {"name": "Ordenada de la tangente", "id": "ordenada"},
            {"name": "Raíz de la tangente", "id": "x_nuevo"},
        ],
        data=[],
        style_table={"marginTop": "20px"},
    ),

    dcc.Interval(id="intervalo", interval=1000, disabled=True),
    dcc.Store(id="contador", data=0),
    dcc.Store(id="x-actual", data=0),
    dcc.Store(id="grafico-datos", data={"tangentes": [], "puntos": [], "raices": []}),
    dcc.Store(id="trazas-tangentes", data=[]),
    html.Div(id='resultado-final'),
    dcc.Graph(id='grafico-funcion', style={'marginTop': '40px'})
])

def f(x):
    return a*x**3 + b*x**2 + c*x + d

def derivada(x):
    return 3*a*x**2 + 2*b*x + c

def ordenada(x):
    return f(x) - derivada(x) * x

@app.callback(
    Output("funcion-math", "children"),
    Output("tabla-newton", "data"),
    Output("contador", "data"),
    Output("intervalo", "disabled"),
    Output("x-actual", "data"),
    Output("resultado-final", "children"),
    Output("grafico-funcion", "figure"),
    Output("grafico-datos", "data"),
    Output("trazas-tangentes", "data"),
    Input("start-btn", "n_clicks"),
    Input("intervalo", "n_intervals"),
    State("input-a", "value"),
    State("input-b", "value"),
    State("input-c", "value"),
    State("input-d", "value"),
    State("input-n", "value"),
    State("input-i", "value"),
    State("tabla-newton", "data"),
    State("contador", "data"),
    State("x-actual", "data"),
    State("grafico-datos", "data"),
    State("trazas-tangentes", "data"),
    prevent_initial_call=True
)
def raiz(start_clicks, n_intervals, a_val, b_val, c_val, d_val, n_val, i_val, tabla, contador, x_actual, data_acumulada, trazas):
    global a, b, c, d, n, i, funcion

    ctx = dash.callback_context
    trigger_id = ctx.triggered[0]['prop_id'].split('.')[0] if ctx.triggered else None

    colores = ['red', 'green', 'orange', 'purple', 'brown', 'cyan', 'magenta', 'gold', 'black']

    if trigger_id == "start-btn":
        a = float(a_val) if a_val is not None else 0
        b = float(b_val) if b_val is not None else 0
        c = float(c_val) if c_val is not None else 0
        d = float(d_val) if d_val is not None else 0
        n = float(n_val) if n_val is not None else 0
        i = int(i_val) if i_val is not None else 0
        funcion = f"{a}x³ {b:+}x² {c:+}x {d:+}".replace("+", " + ").replace("-", " - ")

        return [
            html.Div([f"f(x) = {funcion}"], style={'fontWeight': 'bold'}),
            [],
            0,
            False,
            n,
            "",
            go.Figure(),
            {"tangentes": [], "puntos": [], "raices": []},
            []
        ]

    elif trigger_id == "intervalo":
        if contador >= i:
            return [
                dash.no_update,
                dash.no_update,
                dash.no_update,
                True,
                dash.no_update,
                f"Raíz aproximada: {x_actual} (iteraciones: {i})",
                dash.no_update,
                data_acumulada,
                trazas
            ]

        x = float(x_actual)
        fx = f(x)
        dx = derivada(x)

        if dx == 0:
            return [
                dash.no_update,
                dash.no_update,
                dash.no_update,
                True,
                dash.no_update,
                "Error: Derivada cero",
                dash.no_update,
                data_acumulada,
                trazas
            ]

        x_nuevo = -ordenada(x) / derivada(x)
        # Dominio para graficar f(x)
        x_vals = np.linspace(x - 5, x + 5, 400)
        y_vals = [f(xi) for xi in x_vals]

        # Tangente en x
        tangente_vals = [derivada(x) * (xi - x) + fx for xi in x_vals]
        # Guardar las curvas
        nuevas_tangentes = data_acumulada["tangentes"] + [ [x_vals.tolist(), tangente_vals] ]
        nuevos_puntos = data_acumulada["puntos"] + [ [x, fx] ]
        nuevas_raices = data_acumulada["raices"] + [ [x_nuevo, 0] ]

        # Agregar nueva traza coloreada
        color = colores[contador % len(colores)]
        nueva_traza = go.Scatter(
            x=x_vals,
            y=tangente_vals,
            mode='lines',
            name=f'Tangente {contador + 1}',
            line=dict(color=color, dash='dot')
        )
        nuevas_trazas = trazas + [nueva_traza]

        fig = go.Figure()

        fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', name='f(x)', line=dict(color='blue')))

        for traza in nuevas_trazas:
            fig.add_trace(traza)

        for p in nuevos_puntos:
            fig.add_trace(go.Scatter(x=[p[0]], y=[p[1]], mode='markers', marker=dict(color='red', size=6), name='xₙ'))

        for r in nuevas_raices:
            fig.add_trace(go.Scatter(x=[r[0]], y=[0], mode='markers', marker=dict(color='green', size=8, symbol='x'), name='Raíz tangente'))

        x_combinado = list(x_vals)
        for p in nuevos_puntos:
            x_combinado.append(p[0])
        for r in nuevas_raices:
            x_combinado.append(r[0])
        x_min = min(x_combinado)
        x_max = max(x_combinado)
        x_margin = (x_max - x_min) * 0.1 if x_max != x_min else 1
        y_combinado = y_vals + tangente_vals
        y_min = min(y_combinado)
        y_max = max(y_combinado)

        y_margin = (y_max - y_min) * 0.1
        y_min -= y_margin
        y_max += y_margin

        fig.update_layout(title="Método de Newton-Raphson", xaxis_title="x", yaxis_title="f(x)", height=500, xaxis=dict(range=[x_min - x_margin, x_max + x_margin]), yaxis=dict(range=[y_min, y_max]))

        nueva_fila = {
            "iterador": contador + 1,
            "x": f"{x}",
            "fx": f"{fx}",
            "derivada": f"{dx}",
            "ordenada": f"{ordenada(x)}",
            "x_nuevo": f"{x_nuevo}"
        }

        return [
            dash.no_update,
            tabla + [nueva_fila],
            contador + 1,
            False,
            x_nuevo,
            "",
            fig,
            {
                "tangentes": nuevas_tangentes,
                "puntos": nuevos_puntos,
                "raices": nuevas_raices
            },
            nuevas_trazas
        ]

    return dash.no_update

if __name__ == "__main__":
    app.run(port="8086",debug=False)
````

````python
# Paso 1: Pedir coeficientes
a = float(input("Coeficiente cúbico (a): "))
b = float(input("Coeficiente cuadrático (b): "))
c = float(input("Coeficiente lineal (c): "))
d = float(input("Término independiente (d): "))

# Paso 2: Definir función y derivada
def f(x):
    return a*x**3 + b*x**2 + c*x + d

def f_prime(x):
    return 3*a*x**2 + 2*b*x + c

# Paso 3: Valores iniciales
x0 = float(input("Valor inicial x₀: "))
tolerancia = float(input("Tolerancia (ej. 0.0001): "))
max_iter = int(input("Máximo número de iteraciones: "))

# Paso 4: Iteración de Newton-Raphson
valores = []  # Para tabla
for i in range(max_iter):
    fx = f(x0)
    fpx = f_prime(x0)
    if fpx == 0:
        print("Derivada cero. No se puede continuar.")
        break
    x1 = x0 - fx / fpx
    valores.append((i+1, x0, fx, fpx, x1))
    if abs(x1 - x0) < tolerancia:
        break
    x0 = x1

# Resultado final
print(f"\nRaíz aproximada: {x1:.6f}")
print("\nTabla de iteraciones:")
print("Iter |    x      |   f(x)    |  f'(x)   |   Próximo x")
for v in valores:
    print(f"{v[0]:4} | {v[1]:9.5f} | {v[2]:9.5f} | {v[3]:9.5f} | {v[4]:9.5f}")

````
